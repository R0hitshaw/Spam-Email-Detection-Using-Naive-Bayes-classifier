```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# Download NLTK stopwords (optional, comment out if already downloaded)
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Replace 'your_dataset.csv' with the path to your CSV file containing email text and labels (spam/ham)
data = pd.read_csv('/content/spam.csv')

# Extract email text and labels
emails = data['Message']
labels = data['Category']
```

```python
def preprocess_text(text):

  # Lowercase conversion
  text = text.lower()

  # Remove punctuation (optional)
  # import string
  # punctuation = string.punctuation
  # text = ''.join([c for c in text if c not in punctuation])

  # Remove stopwords (optional)
  stop_words = stopwords.words('english')
  text = ' '.join([word for word in text.split() if word not in stop_words])

  return text

# Apply preprocessing if desired
  emails = emails.apply(preprocess_text)
```

```python
vectorizer = TfidfVectorizer(max_features=2000)  # Adjust max_features as needed
features = vectorizer.fit_transform(emails)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
print (X_train)
print(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)

```

```python
labels = y_test.unique()

# Set the pos_label to one of the labels in the y_test data
pos_label = labels[0]

accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions,pos_label=pos_label)
recall = recall_score(y_test, predictions,pos_label=pos_label)
cm = confusion_matrix(y_test, predictions)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Confusion Matrix:\n{cm}")
```
